# Guia de instalação do JupyterLab no Windows 10

### Passo 1 - Instalar Python

 - Download: https://www.python.org/ftp/python/3.7.9/python-3.7.9-amd64.exe

   Obs: A sugestão é instalar a versão 3.7.9. Site oficial com mais informações sobre a instalação do Python: https://www.python.org/downloads/
   
   - No terminal, execute o comando a seguir para ver a versão do pip instalada:
   
       `pip -V`

### Passo 2 - Instalar JupyterLab

 - Com o Python e o gerenciador de pacotes instalado (pip) podemos fazer a instalação do JupyterLab:
     - No terminal, execute o comando a seguir para instalar o JupyterLab com pip:
  
         `pip install jupyterlab`
         
   Obs: site oficial com mais informações sobre a instalação do Jupyter: https://jupyter.org/install.html 

### Passo 3 - Executar JupyterLab

 - Após a instalação, crie uma pasta onde serão criados os arquivos/notebooks do JupyterLab:
     - No terminal, execute o comando a seguir dentro da pasta criada para iniciar o JuptyerLab:
  
         `jupyter-lab`
         
       O navegador será aberto e o JupyterLab estará pronto para uso.
       
     - Para fechar o programa, abra o terminal novamente e pressione **Ctrl+C** duas vezes.
